## Matplotlib example workflow


```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [2,4,6,8,10]
y = [22,44,66,88,110]

# 2. Setup plot
fig , ax = plt.subplots(figsize = (10,10)) #(width,height)

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel='y-axis')

# 5. Save & show (you save the whole figure)
fig.savefig("./images/sample-plot.png")
```

```python
#making dictionary plot

nut_butter_price = {"almont butter":30,
                    "Peanut butter":40,
                    "Cashew butter":10}
fig,ax = plt.subplots()
ax.bar(nut_butter_price.keys(),nut_butter_price.values())
ax.set(title="Butter store",ylabel="Price ()");
```

### Two options for subplots

```python
#subplot option 1

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                         figsize=(10,5))

ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_price.keys(),nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
#Subplot options 2

fig,ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))
ax[0,0].plot(np.random.random(10))
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_price.keys(),nut_butter_price.values())
ax[1,1].hist(np.random.randn(1000));
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
#PyPlot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
```

```python
# OO method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

# ax.set_xlim(45,100)
```

```python
over_50
```

```python
# OO method for scatter

fig,ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50['age'],
          y=over_50['chol'],
          c=over_50['target'])

# Customize the plot

ax.set(title="Heart Disease and Colesterol Levels",
      xlabel='age',
      ylabel='chol');

#add legend element

ax.legend(*scatter.legend_elements(),title='Target');

# add horizontal line
ax.axhline(over_50['chol'].mean(),linestyle = '--');
```

```python
over_50.head()
```

```python
# Subplot of chol,age,thalach

fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                            figsize=(10,10),
                            sharex=True) 

# Add data in ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);

# Customize ax0
ax0.set(title="Heart Disease and Colesterol Levels",
      ylabel='chol');

# add legend method in ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add meanline in ax0
ax0.axhline(over_50['chol'].mean(),linestyle='--');


# Add data in ax1
scatter = ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target']);

# Customize ax1
ax1.set(title = "Heart Disease and thalach Levels",xlabel='age',ylabel='thalach')

# add legend method ax1
ax1.legend(*scatter.legend_elements(),title="Terget")

# add meanline in ax1

ax1.axhline(over_50['thalach'].mean(),linestyle='--');

# Add the title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight='bold');
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')
# OO method for scatter

fig,ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50['age'],
          y=over_50['chol'],
          c=over_50['target'],
                cmap='summer') # change the color of 

# Customize the plot

ax.set(title="Heart Disease and Colesterol Levels",
      xlabel='age',
      ylabel='chol');

#add legend element

ax.legend(*scatter.legend_elements(),title='Target');

# add horizontal line
ax.axhline(over_50['chol'].mean(),linestyle = '--');
```

```python
# Customizing the ya and x axis limitetions

# Subplot of chol,age,thalach

fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                            figsize=(10,10),
                            sharex=True) 

# Add data in ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter');

# Customize ax0
ax0.set(title="Heart Disease and Colesterol Levels",
      ylabel='chol');

# change the x axis limit
ax0.set_xlim([50,80])

# add legend method in ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add meanline in ax0
ax0.axhline(over_50['chol'].mean(),linestyle='--');


# Add data in ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='winter');

# Customize ax1
ax1.set(title = "Heart Disease and thalach Levels",
        xlabel='age',
        ylabel='thalach')

# change x adn y axis limitation
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# add legend method ax1
ax1.legend(*scatter.legend_elements(),title="Terget")

# add meanline in ax1

ax1.axhline(over_50['thalach'].mean(),linestyle='--');

# Add the title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight='bold');
```
